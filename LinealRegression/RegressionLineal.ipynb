## Desenso del gradiente


```python
import matplotlib.pyplot as plt
from numpy.random import randint
from numpy import linspace
%matplotlib notebook
# Ejemplo del gradiente descendente aplicado a la función y = x^2 + 1
# La ecuación matemática para el gradiente (derivada) es = 2*x
x_inicial = randint(10)
alpha = 0.1
n_iteraciones = 15
iteraciones = []
y = []
x = 16
for i in range(n_iteraciones):
    #print('------------------------')
    #print('iteración ', str(i+1))
    # Calcular gradiente
    gradiente = 2*x
    print("la gradiente es ",gradiente)
    # Actualizar "x" usando gradiente descendente
    x = x - alpha*gradiente
    print("actualizacion de ",x)
    # Almacenar iteración y valor correspondiente
    y.append(x**2+1)
    iteraciones.append(i+1)

    # Imprimir resultados
    #print('x = ', str(x), ', y = ', str(x**2+1))
plt.subplot(1,2,1)
plt.plot(iteraciones,y)
plt.xlabel('Iteración')
plt.ylabel('y')
X = linspace(-5,5,100)
Y = X**2 +1
plt.subplot(1,2,2)
plt.plot(X,Y,0.0,1.0,'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Rgresion lineal

```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def calcular_modelo(w,b,x):
    return w*x+b

def calcular_error(y,y_):
    N = y.shape[0]
    error = np.sum((y-y_)**2)/N
    return error

def gradiente_descendente(w_, b_, alpha, x, y):
    N = x.shape[0]  
    dw = -(2/N)*np.sum(x*(y-(w_*x+b_)))
    db = -(2/N)*np.sum(y-(w_*x+b_))
    w = w_ - alpha*dw
    b = b_ - alpha*db
    return w, b


datos = pd.read_csv('dataset.csv', sep=",", skiprows=32, usecols=[2,3])
datos.plot.scatter(x='Age', y='Systolic blood pressure')
plt.xlabel('Edad (años)')
plt.ylabel('Presión sistólica (mm de Mercurio)')
plt.show()

x = datos['Age'].values
y = datos['Systolic blood pressure'].values

np.random.seed(2)           # Para reproducibilidad del entrenamiento
w = np.random.randn(1)[0]
b = np.random.randn(1)[0]
print("el vector x",x)
print("el vector y",y)
print("el vector w",w)
print("el vector b",b)
print("el vector x",x)
x=-0.1*x
print(x)
alpha = 0.0004
nits = 40000

# Entrenamiento
error = np.zeros((nits,1))
for i in range(nits):
    # Actualizar valor de los pesos usando el gradiente descendente
    [w, b] = gradiente_descendente(w,b,alpha,x,y)
    # Calcular el valor de la predicción
    y_ = calcular_modelo(w,b,x)

    # Actualizar el valor del error
    error[i] = calcular_error(y,y_)

    # Imprimir resultados cada 1000 epochs
    #if (i+1)%1000 == 0:
        #print("Epoch {}".format(i+1))
        #print("    w: {:.1f}".format(w), " b: {:.1f}".format(b))
        #print("    error: {}".format(error[i]))
        #print("=======================================")

# Gráfica de ECM vs iteraciones y de la regresión lineal resultante
plt.subplot(1,2,1)
plt.plot(range(nits),error)
plt.xlabel('epoch')
plt.ylabel('ECM')

y_regr = calcular_modelo(w,b,x)
plt.subplot(1,2,2)
plt.scatter(x,y)
plt.plot(x,y_regr,'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
edad = 90
presion = calcular_modelo(w,b,edad)
```

## Linear Regression simple

hallar la derivada de la funcion
$$ \frac{\partial }{\partial \theta_{j}}J(\theta_{0},\theta_{1})
   = \frac{\partial}{\partial \theta_{j}} \frac{1}{2m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)}) $$
derivadas parciales de la funcion costo o error    
$$\theta_{0}=\theta_{0}-\alpha\frac{1}{m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)})$$

$$\theta_{1}=\theta_{1}-\alpha\frac{1}{m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)})x^{(i)}$$	

      

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import shuffle
import seaborn as sns
%matplotlib inline  

class LinearRegression(object):
    def __init__(self,learningRatio=0.05,numIterations=10000,fitIntercept=True):
        self.learningRatio=learningRatio#parametro de aprendizaje
        self.numIterations=numIterations#numero de iteraciones
        self.fitIntercept=fitIntercept
        self.N=0#cantidad de datos 
        self.X=[]
        self.Y=[]
        self.theta0 = np.random.randn(1)[0]#se genera theta0 y theta1 aleatorios para poder ajustar bien la recta 
        self.theta1 = np.random.randn(1)[0]
    def gradientDesent(self,theta0,theta1):
        dTheta0=(1/self.N)*np.sum(theta0+theta1*self.X-self.Y)
        dTheta1=(1/self.N)*np.sum((theta0+theta1*self.X-self.Y)*self.X)
        theta0=theta0-self.learningRatio*dTheta0
        theta1=theta1-self.learningRatio*dTheta1
        return theta0,theta1
    def fit(self,X,Y):    
        self.X=X
        self.Y=Y
        theta0=self.theta0
        theta1=self.theta1
        self.N=X.shape[0]
        for i in range(self.numIterations):
            [theta0,theta1]=self.gradientDesent(theta0,theta1)
        return theta0,theta1
def readData():
    df = pd.read_csv('Iris.csv')
    df = df.drop(['Id'],axis=1)
    rows = list(range(100,150))
    df = df.drop(df.index[rows])  
    Y = []
    target = df['Species']
    for val in target:
        if(val == 'Iris-setosa'):
            Y.append(0)
        else:
            Y.append(1)
    df = df.drop(['Species'],axis=1)
    X = df.values.tolist()
    X, Y = shuffle(X,Y)
    X=np.array(X)
    Y=X[:,0:1]
    X=X[:,1:2]
    
    return X,Y
def getRect(theta0,theta1,x):
    return theta1*x+theta0
if __name__=='__main__':
    [X,Y]=readData()
    plt.scatter(X,Y)
    lr=LinearRegression()
    theta0=0
    theta1=0
    [theta0,theta1]=lr.fit(X,Y)
    print("los thetas son",theta0,theta1)
    #print(getRect(theta0,theta1,X))
    plt.plot(X,getRect(theta0,theta1,X),'r')
    plt.show()
```

# Multivariable Regression
 
The original, univariate hypothesis function was like:

$h_{\theta}(x)=\theta_{0}+\theta_{1}x$

Para multiples variables <br />
$h_{\theta}(x)=\theta_{0}+\theta_{1}x_{1}+\theta_{2}x_{2}+......+\theta_{n}x_{n}$ <br/>
$x_{0}=1$<br/>
$h_{\theta}(x)=\theta_{0}x_{0}+\theta_{1}x_{1}+\theta_{2}x_{2}+......+\theta_{n}x_{n}$<br/>
$
\vec{\theta}=
\begin{bmatrix}
    \theta_{0}\\
    \theta_{1}\\
    \theta_{2}\\
    \vdots\\
    \theta_{n}
\end{bmatrix}$
$
\vec{x}=
\begin{bmatrix}
    x_{0}\\
    x_{1}\\
    x_{2}\\
    \vdots\\
    x_{n}
\end{bmatrix}$
Esta encuacion se puede simplificar como <br/>
$h_{\theta}(x)=\vec{\theta}^T\vec{x}$
### Desenso del Gradiente version multivariable
funcion costo
$$ J(\vec{\theta})=\frac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)})-y^{(i)}) $$
derivadas parciales con respecto a $\theta_{j}$<br/>
$$repeatUntilConvergence:\{\\
    \theta_{j}:=\theta_{j}-\alpha\frac{\partial}{\partial\theta_{j}}J(\vec{\theta})\\
\}$$
$$ \frac{\partial }{\partial \theta_{j}}J(\vec{\theta})
   = \frac{\partial}{\partial \theta_{j}} \frac{1}{m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)}){x_{j}}^{(i)} $$
  
   $$
   repeat until convergence:\{\\
   \theta_{0}=\theta_{0}-\alpha\frac{1}{m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)}){x_{0}}^{(i)}\\
   \theta_{1}=\theta_{1}-\alpha\frac{1}{m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)}){x_{1}}^{(i)}\\
   \theta_{2}=\theta_{2}-\alpha\frac{1}{m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)}){x_{2}}^{(i)}\\
   \vdots\\
   \}
   $$	
$$\theta_{j}=\theta_{j}-\alpha\frac{1}{m}\sum_{i=1}^{n}(h_{\theta}(x^{(i)})-y^{(i)}){x_{j}}^{(i)}$$
   
   


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import shuffle
import seaborn as sns
%matplotlib inline  

class MultivariateRegression(object):
    def __init__(self,learningRatio=0.05,numIterations=10,fitIntercept=True):
        self.learningRatio=learningRatio#parametro de aprendizaje
        self.numIterations=numIterations#numero de iteraciones
        self.fitIntercept=fitIntercept
        self.N=0#cantidad de datos 
        self.X=[]
        self.Y=[]
        #self.theta0= np.random.randn(1)[0]#se genera theta0 y theta1 aleatorios para poder ajustar bien la recta 
        self.thetaJ=[]
    def gradientDesent(self,thetaJ):
        for i in range(self.X.shape[1]):
            #j=1*2 x=2*100  [1*100] -y=1*100)*x[]
            #thetan=self.thetaJ.T@self.X.T
            #print("hello.... is...is",thetan.shape,thetan)
            dThetaJ=(1/self.N)*np.sum(((self.thetaJ.T@self.X.T)-self.Y.T)@self.X[:,i:i+1])
            self.thetaJ[i]=self.thetaJ[i]-self.learningRatio*dThetaJ
        return thetaJ
    def fit(self,X,Y):    
        self.X=X
        self.Y=Y
        self.N=X.shape[0]
        for i in range(X.shape[1]):
            self.thetaJ.append(np.random.randn(1)[0])
        self.thetaJ=np.array(self.thetaJ)
        self.thetaJ=self.thetaJ[:,np.newaxis]
        print(self.thetaJ.shape)
        for i in range(self.numIterations):
            self.thetaJ=self.gradientDesent(self.thetaJ)
        return self.thetaJ
def readData():
    df = pd.read_csv('Iris.csv')
    df = df.drop(['Id'],axis=1)
    rows = list(range(100,150))
    df = df.drop(df.index[rows])  
    Y = []
    target = df['Species']
    for val in target:
        if(val == 'Iris-setosa'):
            Y.append(0)
        else:
            Y.append(1)
    df = df.drop(['Species'],axis=1)
    X = df.values.tolist()
    X, Y = shuffle(X,Y)
    X=np.array(X)
    #Y=X[:,3:4]
    #X=X[:,:3]
    Y=X[:,0:1]
    X=X[:,1:2]
    return X,Y
def getRect(thetaJ,x):
    return thetaJ[1]*x+thetaJ[0]
if __name__=='__main__':
    [X,Y]=readData()
    plt.scatter(X,Y)
    one=np.full(X.shape[0],1)
    X=np.concatenate((one[:, np.newaxis], X), axis=1)
    lr=MultivariateRegression(numIterations=1000)
    thetaj=[]
    X=X
    thetaj=lr.fit(X,Y)
    plt.plot(X,getRect(thetaj,X),'r')
    plt.show()
    
```
